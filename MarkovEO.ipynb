```python
import numpy as np
from numpy import linalg as LA
import random
nameList = ['iteration 1','iteration 6','iteration 100', 'iteration 1000','iteration 3800','actual pi vector']

def rand_stoch_vect(row, column):
    matrix = np.random.rand(row, column) #set vector size
    v = matrix/matrix.sum(axis=1)[:,None] #ensure vector is stochastic
    v = np.around(v, decimals=3) #limit 3 decimal places
    return v 
def add_stoch_vector(v1, v2): 
    v3 = v1+v2
    v = v3/v3.sum(axis=1)[:,None]
    v = np.around(v, decimals=3)
    return(v)
def subtract_stoch_vectors(v1,v2):
    v3 = abs(v1-v2)
    v = v3/v3.sum(axis=1)[:,None]
    v = np.around(v, decimals=3)
    return(v)
def mult_stoch_vectors(v1,v2): #v1 and v2 must be stochastic
    v3 = v1 * v2
    v = v3/v3.sum(axis=1)[:,None]
    v = np.around(v, decimals=3)
    return(v)
def shuffle(v):
    for n in range(len(v) - 1):
        rnd = random.randint(0, (len(v) - 1))
        val1 = v[rnd]
        val2 = v[rnd - 1]
        v[rnd - 1] = val1
        v[rnd] = val2
    return v


import plotly.graph_objs as go
def vector_plot(tvects,is_vect=True,orig=[0,0,0]):
    """Plot vectors using plotly"""
    if is_vect:
        if not hasattr(orig[0],"__iter__"):
            coords = [[orig,np.sum([orig,v],axis=0)] for v in tvects]
        else:
            coords = [[o,np.sum([o,v],axis=0)] for o,v in zip(orig,tvects)]
    else:
        coords = tvects

    data = []
    for i,c in enumerate(coords):
        X1, Y1, Z1 = zip(c[0])
        X2, Y2, Z2 = zip(c[1])
        vector = go.Scatter3d(x = [X1[0],X2[0]],
                              y = [Y1[0],Y2[0]],
                              z = [Z1[0],Z2[0]],
                              marker = dict(size = [0,5],
                                            color = ['blue'],
                                            line=dict(width=5,
                                                      color='DarkSlateGrey')),
                              name = nameList[i])
        data.append(vector)
    layout = go.Layout(
             margin = dict(l = 4,
                           r = 4,
                           b = 4,
                           t = 4)
                  )
    fig = go.Figure(data=data,layout=layout)
    fig.show()

    
def _flatten(values):
    if isinstance(values, np.ndarray):
        yield values.flatten()
    else:
        for value in values:
            yield from _flatten(value)
def flatten(values):
    # flatten nested lists of np.ndarray to np.ndarray
    return np.concatenate(list(_flatten(values)))

def _unflatten(flat_values, prototype, offset):
    if isinstance(prototype, np.ndarray):
        shape = prototype.shape
        new_offset = offset + np.product(shape)
        value = flat_values[offset:new_offset].reshape(shape)
        return value, new_offset
    else:
        result = []
        for value in prototype:
            value, offset = _unflatten(flat_values, value, offset)
            result.append(value)
        return result, offset

def unflatten(flat_values, prototype):
    # unflatten np.ndarray to nested lists with structure of prototype
    result, offset = _unflatten(flat_values, prototype, 0)
    assert(offset == len(flat_values))
    return result

```
